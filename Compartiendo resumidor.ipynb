````python
from google.colab import drive
import os

drive.mount('/content/drive')

# Directorio donde se encuentran los archivos PDF
directory = '/content/drive/MyDrive/Colab Notebooks/Value investing/'

# Lista para almacenar los nombres de los archivos PDF
pdf_files = []

# Iterar sobre los archivos en el directorio
for filename in os.listdir(directory):
    # Verificar si el archivo tiene la extensión .pdf
    if filename.endswith(".pdf"):
        # Construir la ruta completa al archivo PDF
        pdf_path = os.path.join(directory, filename)
        # Agregar la ruta a la lista de archivos PDF
        pdf_files.append(pdf_path)

# Imprimir la lista de archivos PDF encontrados
print("Archivos PDF encontrados:")
for pdf_path in pdf_files:
    print(pdf_path)
````

````python
from google.colab import drive
from PyPDF2 import PdfReader
import os
# Crear una función para procesar un archivo PDF
def process_pdf(pdf_path):
    pdfFileObject = open(pdf_path, 'rb')
    pdfReader = PdfReader(pdfFileObject)
    text = []

    # Storing the pages in a list
    for i in range(0, len(pdfReader.pages)):
        # creating a page object
        pageObj = pdfReader.pages[i].extract_text()
        pageObj = pageObj.replace('\t\r', '')
        pageObj = pageObj.replace('\xa0', '')
        # extracting text from page
        text.append(pageObj)

    return text

# Procesar cada archivo en la lista pdf_files
for pdf_path in pdf_files:
    text = process_pdf(pdf_path)

    # Haz lo que necesites con el texto procesado, por ejemplo, imprimir la longitud de cada PDF
    print(f"Original Pages ({pdf_path}) = ", len(text))
````

````python
# Merge multiple page - to reduce API Calls
def join_elements(lst, chars_per_element):
    new_lst = []
    for i in range(0, len(lst), chars_per_element):
        new_lst.append(''.join(lst[i:i+chars_per_element]))
    return new_lst

# Option to keep x elements per list element

#new_text = join_elements(text, 3)

#print(f"Compressed Pages = ",len(new_text))
#print(f"Compressed Pages ({pdf_path}) = ", len(new_text))

# Procesar cada archivo en la lista pdf_files
for pdf_path in pdf_files:
    text = process_pdf(pdf_path)
    new_text = join_elements(text, 3)
    # Haz lo que necesites con el texto procesado, por ejemplo, imprimir la longitud de cada PDF
    print(f"Original Pages ({pdf_path}) = ", len(text))
    print(f"Compressed Pages ({pdf_path}) = ", len(new_text))
````
