```python
#especificamos la carpeta donde tenemos los .txt y con ello comenzamos a unificarlos

from google.colab import drive
import os

drive.mount('/content/drive')

# Directorio donde se encuentran los archivos de texto
directory = '/content/drive/MyDrive/Colab Notebooks/pdf negocios y psicologia pdf part 1/'
directory_name = os.path.split(directory.rstrip(os.path.sep))[-1]

# Lista para almacenar los nombres de los archivos de texto
text_files = []

# Función para procesar un archivo de texto
def process_text(text_path):
    with open(text_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Iterar sobre los archivos en el directorio
for filename in os.listdir(directory):
    # Verificar si el archivo tiene la extensión .txt
    if filename.endswith(".txt"):
        # Construir la ruta completa al archivo de texto
        text_path = os.path.join(directory, filename)
        # Agregar la ruta a la lista de archivos de texto
        text_files.append(text_path)

# Directorio donde se guardará el archivo de texto combinado
#output_directory = '/content/drive/MyDrive/Colab Notebooks/epub aprender a aprender/'
output_directory = directory

# Nombre del archivo de texto combinado
combined_text_file = f'combined_text_de_[{directory_name}].txt'

# Procesar cada archivo de texto en la lista text_files
combined_text = ""
for text_path in text_files:
    text = process_text(text_path)
    #Ahora extraemos el nombre del archivo de la ruta PDF
    text_name = os.path.basename(text_path)
    combined_text += "\n\n"+ text_name + "\n\n"+ text + "\n\n\n\n" + "    END   " + "\n"

# Guardar el contenido combinado en un nuevo archivo de texto
output_path = os.path.join(output_directory, combined_text_file)
with open(output_path, 'w', encoding='utf-8') as output_file:
    output_file.write(combined_text)

print(f"Contenido de todos los archivos de texto combinado y guardado en {output_path}")
```

```python
#ahora vamos a limpiar de los errores de la IA  como por ejemeplo la repeticion de los promts
# en este caso unicamente para el archivo este el combined_text"
#----------------------------------------------------------------------------------------------

# Abrir el archivo de texto en modo lectura
with open(output_path, 'r', encoding='utf-8') as file:
    contenido = file.read()

# Texto que deseas eliminar
texto_a_eliminar = "The summarizer's task is to summarize the content of the text he reads, without lying or inventing, ensuring clarity and accessibility for any reader. He condenses the text in less than 150 words per page while maintaining precision and coherence. He tries to capture the main ideas and key concepts, providing explanations and clear examples to help understand. His objective is to facilitate the understanding and acquisition of knowledge, making the content accessible to a broad audience. In addition, he identifies and highlights any connection or pattern between ideas on the pages."

# Eliminar el texto especificado
contenido_modificado = contenido.replace(texto_a_eliminar, "")

# Abrir el archivo de texto en modo escritura y escribir el contenido modificado
with open(output_path, 'w', encoding='utf-8') as file:
    file.write(contenido_modificado)

print("Archivo modificado exitosamente.")

#funciono

```

```python
#acá creamos un documento word

from docx import Document

# Función para cargar el contenido de un archivo de texto
def cargar_contenido(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        return f.read()

# Directorio donde se encuentra el archivo de texto combinado
archivo_texto = output_path

# Nombre del archivo Word
archivo_word = output_path + '.docx'

# Cargar el contenido del archivo de texto
contenido = cargar_contenido(archivo_texto)

# Crear un objeto Document de Word
doc = Document()

# Agregar el contenido al documento de Word
doc.add_paragraph(contenido)

# Guardar el documento de Word
doc.save(archivo_word)

print(f"Documento Word creado y guardado en {archivo_word}")
```
